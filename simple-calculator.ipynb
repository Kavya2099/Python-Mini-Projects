```python
print("Simple Calculator")

while(True):
    print("********")
    print("1.Add \n2.Subtract \n3.Multiply \n4.Divide")
    print("********")
    a= float(input("Enter first number: "))
    b= float(input("Enter second number: "))
    c=input("Enter the arthimetic operation number you choose to perform: ")
    print("********")

    if c=='1':
        print("Result: ",a+b)
    elif c=='2':
        print("Result: ",a-b)
    elif c=='3':
        print("Result: ",a*b)
    elif c=='4':
        print("Result: ",a/b)
    else:
        print("You have provided invalid choice. Please try again")

    loop= input("Do you wish to continue: Y/N")
    if(loop=='N' or loop=='n'):
        break

```
